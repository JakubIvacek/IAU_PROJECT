#  *(C-1b) Vychýlené hodnoty (outlier detection), vyskúšajte riešiť problém min. 2 technikami*

```python
def identify_outliers(a):
    lower = a.quantile(0.25) - 1.5 * stats.iqr(a)
    upper = a.quantile(0.75) + 1.5 * stats.iqr(a)

    return a[(a > upper) | (a < lower)]
```

### p.android.settings
Ako sme zistili pri analýze jednotlivých atribútov p.android settings má vela outlierov


```python
pa_settings_out = identify_outliers(processes['p.android.settings'])
pa_settings_out
```

#### Funkcia na identifikovanie vychýlených hodnôt z cvicenia

### Odstránenie vychýlených alebo odľahlých pozorovaní

```python
pa_settings_cleared = processes.drop(index=pa_settings_out.index)
pa_settings_cleared
```

#### Ocisteny dataset processes o outliery som si uložil do pa_settings_cleared

###  nahradenie vychýlenej hodnoty hraničnými hodnotami rozdelenia (napr. 5%, 95%)


```python
lower_bound = processes['p.android.settings'].quantile(0.05)
upper_bound = processes['p.android.settings'].quantile(0.95)

processes['p.android.settings'] = np.where(
    processes['p.android.settings'] < lower_bound, lower_bound,
    np.where(processes['p.android.settings'] > upper_bound, upper_bound, processes['p.android.settings'])
)
```
