```python
connections.select_dtypes(include='number').corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(connections.select_dtypes(include='number').corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=connections, y='c.android.youtube', x='mwra')
corr = connections['c.android.youtube'].corr(connections['mwra'])
plt.title(f'Correlation: {corr:.2f} Mierne pozitívna corelácia')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=connections, y='c.katana', x='mwra')
corr = connections['c.katana'].corr(connections['mwra'])
plt.title(f'Correlation: {corr:.2f} Mierna negatívna korelácia')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(data=connections, y='c.katana', x='c.android.youtube', scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
corr = connections['c.katana'].corr(connections['c.android.youtube'])
plt.title(f'Correlation: {corr:.2f} Mierna negatívna korelácia')
plt.show()
```

```python
# Slabá pozitívna korelácia
plt.figure(figsize=(8, 6))
sns.regplot(data=connections, y='c.android.gm', x='c.android.youtube', scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
corr = connections['c.android.gm'].corr(connections['c.android.youtube'])
plt.title(f'Correlation: {corr:.2f} Slabá pozitívna korelácia')
plt.show()
```
